# Context Manager

- Context manager is used to manage resources automatically
- It ensures the resource is properly opened and closed properly

```python
# Exampel without context manager
f = open("/content/sample_data/README.md", "r")
data = f.read()
print(data)
f.close()
```

```python
# Example with Context manager
with open("/content/sample_data/README.md", "r") as f:
  data = f.read()
  print(data)
```

```python
# Creating numpy one dimentional array
arr = np.array([1,2,3,4,5])
print(arr)
print(type(arr)) # Type of the array
print(arr.ndim) # Dimension of the array
```

# Performance between numpy and lists

```python
import time

lst = list(range(1000000))
start = time.time()
[x*x for x in lst]
end = time.time()
print("Time taken by list: ", (end-start))

arr = np.array(lst)
start = time.time()
arr = arr*arr
end = time.time()
print("Time taken by array: ", (end-start))
```

# Pandas Series

- One dimensional data structure. Like one column in excel

```python
s = pd.Series([10,20,30,40,50])
print(s)
print(type(s))
```

```python
# Series with Custom index
s = pd.Series([10,20,30,40,50], index=['a','b','c','e','f'])
print(s.loc['a']) # Access by label -> use iloc
print(s.iloc[1]) # Access by position -> use iloc
```

# Series attributes
- s.values
- s.indedx
- s.dtype

```python
# Series attributes
print(s.values) # Values in the format of array
print(s.index) # Prints indexes
print(s.dtype)
```

```python
# Arthemetic operation on series
s = pd.Series([10,20,30])
s + 5
```

```python
# Filtering Series

s = pd.Series([10,20,30,40,50])
s[s>30]
```

```python
# Apply funciton
s = pd.Series([10,20,30,40,50])
s.apply(lambda x: x*x)
```

```python
# Missing values in series
s = pd.Series([10,20,30,40,50, None])
s.isna()
```

```python
s = pd.Series([10,20,30,40,50, None])
s.fillna(0)
```

```python
# Droppint the values
s = pd.Series([10,20,30,40,50, None])
s.dropna()
```

```python
# Fill with a average value
s = pd.Series([10,20,30,40,50, None]) # Fill the None with mean
print(s)
s.fillna(s.mean())
```

# Pandas DataFrame

- A data frame is a table with rows and columns. Note series is one dimensional

```python
data = {
    'Name' : ['Alice','Bob','Charlie'],
    'Age' : [25,30,35],
    'City' : ['New York','San Francisco','Los Angeles']
}

df = pd.DataFrame(data)
df
```

```python
# Give names to indexes

df = pd.DataFrame(data, index=["emp1","emp2","emp3"])
df
```

```python
# List of dictionaries
data = [
    { 'Name' : 'Alice', 'Age' : 25, 'City' : 'New York'},
    { 'Name' : 'Bob', 'Age' : 30, 'City' : 'San Francisco'},
    { 'Name' : 'Charlie', 'Age' : 35, 'City' : 'Los Angeles'}
]
df = pd.DataFrame(data)
print(df)
```

```python
# From a dictionariy of series
data = {
    'Name' : pd.Series(['Alice','Bob','Charlie']),
    'Age' : pd.Series([25,30,35]),
    'City' : pd.Series(['New York','San Francisco','Los Angeles'])
}

df = pd.DataFrame(data)
df
```
