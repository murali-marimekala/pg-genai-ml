# List Comprehension



```python
fruits = ["apple","banana","cherry"]
lst = []
for x in fruits:
  if "a" in x:
    lst.append(x)
print(lst)
```

```python
[x for x in fruits if "a" in x]
```

```python
# Syntax of list compehension

# newlist. [expression for item in iterable if condition == True]
```

```python
# Get list of even number greater than 4 from list
lst = [1,2,4,3,4,5,63,8,16]
a=[]
for x in lst:
  if x%2==0 and x >4:
    a.append(x)
print(a)
```

```python
# Getting above result using List Comprehension
lst = [34,2,4,5,7,7,8,16]
[x for x in lst if x%2 ==0 and x>4 ]

a1 = [x if x>4 else x%2==0 for  x in lst]
print(a1)
```

```python
lst = [1,2,3,4,5,6,7,8]
a1 = ["Even" if x%2 == 0 else "Odd" for x in lst]
print(a1)
```

```python
#Nested if conditions in list comprehension
# Create a single list with all elements in the nested list
lst = [[1,2],[3,4],[5,6]]
lst1 = [x for n in lst for x in n]
print(lst1)
```

```python
# Building a function that can take any number of kids name
# from each household during the survey

def cum_sum(*num1):
  sum = 0
  for i in num1:
    sum = sum + i
  return sum

cum_sum(1,2,3,4)
```

# Lambda Functiona

- A lambda function is a small anonymous fucntion
- It can have any number of arguments but only one expression
- the expression is evaluated and returned
- Its good for performing short operations / data manipulations
- Using the lambda funciton can sometime reduce the readabiliyt of code.

```python
# syntax of lambda function
lambda arguments : expression
```

```python
square = lambda X: X ** 2
print(square(5))
```

```python
# Adding two number
add = lambda x,y : x + y
print(add(2,3))
```

# Map function
- map() is used to apply a function to each element of an iterable


```python
# Syntax of map
# map(funciton , iterable)

# The map function in python is a built-in function that applies a given funciton to each item of an iterable
# Z (such as list , tuple or set)
# Returns a map object (an iterator) containing the results

numbers = [1,2,3,4]
squared = map(lambda x: x**2, numbers)
print(list(squared)) #squared return above is an object. So we need to convert to list
```

```python
# How to get the same result without lambda function
numbers = [1,2,3,4]
def squared(x):
  return x**2

squared_numbers = map(squared, numbers)
print(list(squared_numbers))
```

# Python Filter

- filter() is used to select elements for an iterable based on a condition


```python
nums = [1,2,3,4,5,6]
result = filter(lambda x: x%2==0, nums)
print(list(result))

result = lambda x: x%2==0
print(result([x for i in nums]))
```

# Generator function
- Special function is used when we want to generate one value of time instead of returnign everything at once.
- Saves memory
- Useful for large data or inifinite sequneces
- Values are produced only when needed
- Instead of return we return yield

```python
def numbers():
  yield 1
  yield 2
  yield 3

for n in numbers():
  print(n)
```

```python
val = input("Enter your number")
print(val)
```

```python
def factorial(n):
  fact = 1
  for i in range(1, n+1):
    fact = fact * i
  return fact

val = int(input("Enter your number"))
print("Factorial: ", factorial(val))
```
