# List Comprehension



```python
fruits = ["apple","banana","cherry"]
lst = []
for x in fruits:
  if "a" in x:
    lst.append(x)
print(lst)
```

```python
[x for x in fruits if "a" in x]
```

```python
# Syntax of list compehension

# newlist. [expression for item in iterable if condition == True]
```

```python
# Get list of even number greater than 4 from list
lst = [1,2,4,3,4,5,63,8,16]
a=[]
for x in lst:
  if x%2==0 and x >4:
    a.append(x)
print(a)
```

```python
# Getting above result using List Comprehension
lst = [34,2,4,5,7,7,8,16]
[x for x in lst if x%2 ==0 and x>4 ]

a1 = [x if x>4 else x%2==0 for  x in lst]
print(a1)
```

```python
lst = [1,2,3,4,5,6,7,8]
a1 = ["Even" if x%2 == 0 else "Odd" for x in lst]
print(a1)
```

```python
#Nested if conditions in list comprehension
# Create a single list with all elements in the nested list
lst = [[1,2],[3,4],[5,6]]
lst1 = [x for n in lst for x in n]
print(lst1)
```

# Python - Function

- Function is a re-usuable block of code that performs specific task
- Use keyword def
  - def function_name(param1, param2, ...)
      line1
      line2
      return # Return is the last statement in the function



```python
#Create a fucntion
def my_function():
  print("Hello from a function")
  print("Good morning")
print("Outside the function")
```

```python
# Calling the funciton
my_function()
```

```python
# Arguments/Parameters: Information can be passed into functions as arguments
def PizzaMachine(veggie, sauce):
  print(veggie + " " + "Pizza" + " " + sauce)

PizzaMachine("Pepporoni","Seracha")
```

# Positional Arguements
- Pass the argument in the right order

```python
def greeting(fname, lname):
  print("Hello" + " " + fname)

greeting("Murali", "Krishna")
```

# Keyword Arguments
- You can also send arguments with the key = value syntax


```python
def greeting(fname, lname):
  print("Hello" + " " + fname + " " + lname)

greeting(fname = "Murali", lname="krishna")
```

```python
def diffthis(num1, num2):
  print(num1 - num2)

diffthis(2,10)
```

```python
# Arbitrary Argument - Some function can have different number of arguement
# as it cant be predicted

def my_function(*kids):
  print("The youngest child is " + kids[2])

my_function("Emil", "Tobias", "Rana")
```

```python
# Building a function that can take any number of kids name
# from each household during the survey

def cum_sum(*num1):
  sum = 0
  for i in num1:
    sum = sum + i
  return sum

cum_sum(1,2,3,4)
```

```python
cum_sum(2,3,4)
```

```python
# Building a function that can take any number of kids from each household during the survey

def polioDrive(*kids):
  print(kids)

polioDrive("Murali", "Krishna", "Raj")
```

```python
# Arbitrary argument and positional argument
def my_function(food, *kids):
  print("Kid name is",kids)
  print("I like " + food)

my_function("Biryani","Rana")
```

```python
# Arbitrary Keyword Argueents, **kwargs

def my_function(**kids):
  print(kids)

my_function(kid1 = "Murali", kid2 = "Krishna", house_number = 2, doctor="Xyz")
```

```python
def newfunc(*marks, **studentdetail):
  print("Positional Args:", marks)
  print("Keyword Args", studentdetail)

newfunc(50,56,34,name = "Murali",class_name = "12th")
```
