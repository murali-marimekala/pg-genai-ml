# Variables

```python
# Variables in Python is created when you asisgn a value to it. Python is a dynamically typed language, which means you don't need to declare the type of variable explicitly.
x = 5
y = "John"

print(x)
print(y)
    
```

```python
# You can change the value of a variable, and it can even change its type.
x = 10
x = "Hello"
print(x)
```

```python
# Casting in Python is the process of converting a variable from one type to another. 
# You can use built-in functions like str(), int(), and float() to perform casting.
x = str(3)    # x will be '3
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0
print(x)
print(y)
print(z)
```

```python
# Type() function is used to determine the type of a variable. 
x = 5
y = 10.5
print(type(x))
print(type(y))
```

```python
#Single and double quotes can be used to create string literals in Python. 
# Both are functionally equivalent, and you can use either based on your preference or to avoid the need for escaping characters.
str = "Hello World"
print(str)  

str2 = 'Hello World'
print(str2)
```

```python
# a will not overwrite A
a=10
A="Rana"
print(A)
```

# Variable names

```python
# Variable names are case-sensitive in Python. This means that a variable named "a" is different from a variable named "A". 

var1 = 10
var_1 = 20
_var_1 = 30
VAR1=40
print(var1)
print(var_1)
print(_var_1)
print(VAR1)
```

```python
# Invalid variable names
var-1 = 50
var 1 = 60
1var=70
```

# Multi Words Variable Names

```python
#Camel case
myVariableName = "Rana"

#Pascal case
MyVariableName = "Rana"

#Snake case
my_variable_name = "Rana"

print(myVariableName)
print(MyVariableName)
print(my_variable_name) 
```

# Many Values to Multiple Variables

```python
# One value to multiple variables
x, y, z = "Orange", 5,  6.8
print(x)
print(y)
print(z)    
```

# One Value to Multiple Variables

```python
# One value to multiple variables
x=y=z=10
print(x)
print(y)
print(z)
```

# Unpack a Collection

```python
# Unpacking a collection
fruits = ["Banana", "Strawberry", "Craneberry"]
x, y, z = fruits
print(x)
print(y)
print(z)
```
