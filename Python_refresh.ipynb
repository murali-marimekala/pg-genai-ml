# Global Variable

```python
#Variables that are created outside a function are known as global variables

x = "awesome"

def myfun():
  print("Python is " + x)

myfun()
```

```python
# If you create a variable with the same name inside a function , this variable wilil be local and can be only used inside the fucntion.

x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x) # Displays fantastic the local variable value

myfunc()

print("Python is " + x) # Displays awesome the global variable value
```

# The Global Keyword

```python
# If you use global keyword, the variable belogs to the global scope

def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is "+x)
```

```python
# Use global keyword if you want ot change a global vairable inside a function

x = "awesome"

def myfunc():
  global x
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x) # global value of x "awesome" is overwritten by "fantastic"
```

# Python - Slicing Strings



```python
b = "Slicing a string"
print(b[2:5]) # Prints from index 2 (which starts with 0) to 4 only. Not 5

#Slice from the start
print(b[:5]) #Print from 0 to 4

# Slice to the end
print(b[2:]) #Print from 2 till end

b = "Hello, World!"
#Negative indexing
print(b[-5:-2]) #Print from -5 to -2
```
