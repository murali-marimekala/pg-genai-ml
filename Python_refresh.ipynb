# Python Operator Precedence

```python
#Precendence order
# First precedence is ()
a = (10 + 5) * 2 # Not evaulated as 10 + (5 * 2)
print(a)
```

```python
# Second is Exponentiation
a = 2 ** 3 ** 2 # 2 ** (3 ** 2)
print(a)
```

```python
# Operator Associativity in Python
# --------------------------------
# Only three categories of operators in Python are right-associative:
#   1. Exponentiation (**)
#   2. Unary operators: +x, -x, ~x
#   3. Walrus operator (:=)
#
# All other binary operators (arithmetic, bitwise, boolean, etc.) are left-associative.
#
# Definitions:
#   Right associative:  a op b op c    a op (b op c)
#   Left associative:   a op b op c    (a op b) op c
```

```python
# Mulitplication, division, floow division and modulus
a = 10 * 2
b = 10 / 2
c = 10 // 2
d = 10 % 2
print(a, b, c, d)

a = 10 * 2 + 10 / 5 # Evaluates as 10 * 2 + (10/5)
print(a)

b = 10 * 10 // 2 + 10 % 2 # Evaluates as 10 * (10 // 2) + (10 % 2)
print(b)
```

```python
# Addition and substration
a = 10 + 5 - 3 # Evaluates as 10 + (5-3) note apart from exponential, unary and walrus rest of operators left associative
print(a)
```

```python
# To append elements from another list to the current list, use extend() method

thelist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya"]
thelist.extend(tropical)
print(thelist)

# We can add tuple items to end of list using extend
thistuple = ("kiwi","strawberr")
thelist.extend(thistuple)
print(thelist)
```

# List Comprehension

- Offers shorter syntax


```python
fruits = ["apple","banana","Cherry","mango","kiwi"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)

print(newlist)
```

```python
# Now lets try List comprehension
# Syntax is newlist = [expression for item in iterable if condition == True]
fruits = ["apple","banana","Cherry","mango","kiwi"]
[x for x in fruits if "a" in x]
```

# Condition

- Codition is like filter that only accepts the items that evaluates to True

```python
newlist = [x for x in fruits if x != "apple"]
print(newlist)
```

# Iterable
- The iterable can be any iterable object, list a list, tuple, set etc

```python
newlist = [x for x in range(10)]
print(newlist)

[x for x in range(10) if x < 5]
```

# Expression

```python
newlist = [x.upper() for x in fruits]
print(newlist)

newlist = ["hello" for x in fruits]
print(newlist)
```

# Python - Sort Lists

```python
# Sort List Alphanumerically

# List objects have a sort() method that will sort the list alpanumerically , ascending by default

thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort()
print(thislist)

thislist = [100, 50, 65, 82,23]
thislist.sort()
print(thislist)
```

```python
# Sort asending
thelist = [23,5,2,4,22,32]
thelist.sort(reverse = True)
print(thelist)
```

# Customize the sort Function

```python
def myfunc(n):
  return abs(n - 50)

thislist = [100, 50, 65, 82,23]
thislist.sort(key = myfunc)
print(thislist)
```

# Case Insensitive Sort

- By default sort function is case sensitive, resulting in all capital letters being sorted before lower case letters.

```python
thislist = ["banana","Orange","Kiwi","Cherry"]
thislist.sort()
print(thislist)
```

```python
# Use built-in function as key functions when sorting a list
thislist = ["banana","Orange","Kiwi","Cherry"]
thislist.sort(key = str.lower) # Case sensitive sort function
print(thislist)
```

# Reverse Order

- revese() method reverses current sorting order of the elements

```python
# Revese the list itemsthislist = ["banana", "Orange","Kiwi","cherry"]
thislist.reverse()
print(thislist)
```

# Python - Copy Lists

1.   You cannot copy a list simply by typing list1 = list2 because list1 willl be a reference to list2
2. Any Change made in list2 will be made automatically on list2



```python
thislist = ["apple", "banana", "cherry"]
mylist = thislist.copy()
print(mylist)
```

# Use the list() method

- Another way to use is built-in list() method

```python
thislist = ["apple", "banana", "cherry"]
mylist = list(thislist)
print(mylist)
```

# Use the Slice Operator

- Another alternative is to use the : (slice) operator

```python
thislist = ["a","b","c","d"]
mylist = thislist[:]
print(mylist)
```
